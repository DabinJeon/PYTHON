```python
#범주형 변수가 주어졌을때 이항변수화(OneHotEncoder)
#남=>1,여=>2, 연령: 10대=>1, 20대=>2, ... 
#등급:S=>1, A=>2, ...
#ex)
#성별(남(0)/여(1)) => 원핫인코딩 => 
#0=>10, 1=>01
#연령대(10대(0),20대(1),30대(2),40대(3))
#0=>1000, 1=>0100, 2=>0010, 3=>0001
```

```python
from sklearn.preprocessing import OneHotEncoder
data=np.array(
#[성별, 연령대, 등급]
[[0,0,0],
 [0,1,1],
 [0,2,2],
 [1,0,3],
 [1,1,4]])
data

```

```python
ohe=OneHotEncoder()
ohe.fit(data)#data를 원핫인코더(ohe)를 이용하여 변환(fit)한다
```

```python
ohe.active_features_
#[남,여,20대,30대,40대, S,A,B,C,D]
```

```python
ohe.n_values_ 
#성별:2, 연령:3, 등급:5    범주의 개수를 출력
```

```python
#여성, 30대, S등급
mydata=np.array([[1,1,0]])
ohe.transform(mydata).toarray()
#1=>01, 1=>010, 0=>10000
ohe.feature_indices_ #0이상2미만(성별), 2이상5미만(연령), 5이상10미만(등급)
```

```python
#np.linspace(0,1,11)
bins=np.linspace(df.A1.min(), df.A1.max(), 4)
```

```python
bins
```

```python
df['A1_bin']=np.digitize(df['A1'], bins)
```

```python
df
```

```python
df.groupby('A1_bin')['A1'].size()
df.groupby('A1_bin')['A1'].median()
df.groupby('A1_bin')['A1'].mean()
df.groupby('A1_bin')['A1'].std()
df.groupby('A1_bin')['A1'].var()
df.groupby('A1_bin')['A2'].value_counts()
```

```python
df['hl']=np.where(df['A1']>=df['A1'].mean(),'high','low')
df
df.groupby('hl')['A1'].size()
df.groupby('hl')['A1'].std()
```
